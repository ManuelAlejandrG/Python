## Dataset de las casas de boston

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```
# Cargamos la libreria
boston = load_boston()
```

 Descripci√≥n del dataset

```
boston.DESCR
```

Con el numero medio de habitaciones vamos a predecir el precio de la vivienda

```
X =np.array( boston.data[:,5]) # columna asociada al numero medio de habitaciones por hogares de cada barrio

y = np.array(boston.target) # precio por hogares en cada barrio
```

```
# Grafica de numero medio de habitaciones vs precio por hogares

plt.scatter(X,y, alpha=0.4)
plt.title("habitaciones vs precio")
plt.xlabel("Numero medio de habitaciones")
plt.ylabel("precio medio por horages")
plt.show()
```

```
# agregamos columnas de 1's para t√©rmino independiente
x = np.array([np.ones(506),X]).T
x.shape
```

La f√≥rmula para el m√≠nimo cuadrado es:

$$\beta = (X^T X)^{-1}X^t Y$$

```

beta = np.linalg.inv(x.T @ x) @x.T@y

beta
```

```


plt.plot([4,9],[beta[0]+beta[1]*4,beta[0]+beta[1]*9], label="modelo", color="red")
plt.scatter(X,y, alpha=0.4, label="valores")
plt.title("habitaciones vs precio")
plt.xlabel("N√∫mero medio de habitaciones")
plt.ylabel("precio medio por horages")
plt.legend()
plt.show()
```
